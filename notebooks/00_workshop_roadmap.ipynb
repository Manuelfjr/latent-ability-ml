# Bristol Workshop Roadmap

This workshop is organized as a 3-hour progression from unsupervised evaluation to binary IRT, and then to Beta4-IRT and CLAIRE.


## Suggested Session Timing

Each section follows the same rhythm:

- **15 minutes** of theory and motivation
- **15 minutes** of guided hands-on explanation
- **30 minutes** for participants to solve the activity notebook

Total: **3 hours**.


## Notebook Sequence

1. `01_00_unsupervised_evaluation_toy_problems.ipynb`  
   Guided notebook for Section 1. Introduces toy clustering datasets, compares clustering models, and motivates the idea that some instances are systematically harder than others.
2. `01_01_activities.ipynb`  
   Activity notebook for Section 1. Participants create or adapt a clustering scenario and inspect how aggregate metrics can hide instance-level structure.
3. `01_99_answer.ipynb`  
   One possible worked solution for the Section 1 activity.
4. `02_00_binary_irt_and_2pl.ipynb`  
   Guided notebook for Section 2. Covers binary IRT, 1PL intuition, 2PL-IRT, and reading ICCs through toy item banks.
5. `02_01_activities.ipynb`  
   Activity notebook for Section 2. Participants manipulate difficulty and discrimination values and interpret the resulting ICCs.
6. `02_99_answer.ipynb`  
   One possible worked solution for the Section 2 activity.
7. `03_00_beta4_and_claire.ipynb`  
   Guided notebook for Section 3. Connects the workshop to the research context: Beta4-IRT, CLAIRE, agreement-based response matrices, abilities, difficulties, and discriminations.
8. `03_01_extra_negative_disc.ipynb`  
   Extra analysis notebook. It was created to investigate, in a more controlled way, how negative discrimination can emerge when the agreement structure is directly perturbed.
9. `03_02_activities.ipynb`  
   Activity notebook for Section 3. Participants generate a synthetic dataset with medium to high variability, build the CLAIRE response matrix, train Beta4-IRT, inspect abilities and item parameters, and generate ICCs.
10. `03_99_answer.ipynb`  
    One possible worked solution for the Section 3 activity.


## Teaching Arc

The workshop is designed as a progression:

- Section 1 motivates the problem: aggregate evaluation in unsupervised learning can hide instance-level variability and treat all instances as equally difficult.
- Section 2 introduces IRT as a latent-variable language for ability, difficulty, and discrimination.
- Section 3 presents Beta4-IRT and connects it to CLAIRE, where model agreement generates the response matrix used for analysis.


## Practical Notes

- All notebooks use toy problems so they can be understood quickly in class.
- The activity notebooks are intentionally incomplete and include prompts for discussion.
- The `*_99_answer.ipynb` notebooks can be used after the participant work to consolidate the discussion.
- `03_01_extra_negative_disc.ipynb` is an extra analytical notebook, not the default participant activity.
- Shared helper code lives mainly in `utils/handson.py`, `utils/transform.py`, and `notebooks/nb_utils.py`.
