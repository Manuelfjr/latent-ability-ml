# Bristol CDT Workshop Roadmap

This repository supports a hands-on session on **supervised evaluation**, **binary IRT**, and **Beta4/CLAIRE** for students at the University of Bristol CDT.

The guiding idea is simple: aggregate metrics are useful, but they hide how model performance changes across easy and hard instances.


## Session Structure

The workshop is organized into three teaching blocks. Each block follows the same rhythm:

- **15 minutes** of theory
- **15 minutes** of guided practical explanation
- **30 minutes** of student activity

This timing keeps the narrative consistent and gives the students enough time to explore the toy problems themselves.


## Notebook Sequence

1. `01_supervised_evaluation_toy_problems.ipynb`
   This notebook shows why standard supervised metrics can be misleading when examples are not equally difficult. We compare multiple models, not only one, and inspect hard instances directly.
2. `02_irt_beta4_icc.ipynb`
   This notebook introduces classical binary IRT, explains ability, difficulty, and discrimination, and uses ICCs to connect latent ability to supervised evaluation.
3. `03_claire_latent_ability.ipynb`
   This notebook moves to the Beta4 perspective and positions CLAIRE as the latent-ability-aware method proposed for this setting.


## Teaching Goals By Section

### 1. Supervised evaluation of methods
- Show that accuracy alone may hide important failures.
- Compare models on balanced and imbalanced toy datasets.
- Motivate the idea that some instances are systematically harder than others.

### 2. Classical binary IRT
- Introduce the meaning of latent ability, item difficulty, and discrimination.
- Use ICCs to show how the probability of success changes with ability.
- Bridge students from educational testing language to model evaluation language.

### 3. Beta4-IRT and CLAIRE
- Show how Beta4 offers a richer response curve family.
- Use item difficulty and discrimination parameters in a toy latent-ability setting.
- Explain where CLAIRE fits in a latent-aware evaluation pipeline.


## Activity Design Principles

Each notebook ends with one short activity set. The activities are intentionally designed around toy problems so that students can focus on interpretation rather than implementation details.

The expected outputs are:

- one metric comparison or ranking;
- one interpretation of hard items or difficult instances;
- one short discussion connecting standard evaluation to latent ability.


## Presenter Checklist

Before the trip, the main development tasks are:

- polish the toy examples and make sure the plots run cleanly;
- decide how formal the CLAIRE definition should be for the CDT audience;
- confirm whether you want to mention parameter estimation or only interpretation;
- prepare one or two oral examples from your own work to connect the notebooks to the research motivation.
