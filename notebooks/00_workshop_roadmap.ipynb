# Bristol Workshop Roadmap

This repository is now organized as a 3-hour workshop with three sections, each one split into a guided notebook and an activity notebook.


## Suggested Session Timing

Each section follows the same rhythm:

- **15 minutes** of theory and motivation
- **15 minutes** of guided hands-on explanation
- **30 minutes** for participants to solve the activity notebook

Total: **3 hours**.


## Notebook Sequence

1. `01_supervised_evaluation_toy_problems.ipynb`
   Standard supervised evaluation and why equal weighting of instances is limiting.
2. `02_supervised_evaluation_activities.ipynb`
   Activity notebook for Section 1.
3. `03_binary_irt_and_2pl.ipynb`
   Binary IRT, 1PL intuition, and 2PL item characteristic curves.
4. `04_irt_activities.ipynb`
   Activity notebook for Section 2.
5. `05_beta4_and_claire.ipynb`
   Beta4-IRT, bounded latent variables, and the CLAIRE workflow.
6. `06_beta4_and_claire_activities.ipynb`
   Activity notebook for Section 3.


## Teaching Arc

The workshop is designed as a progression:

- Section 1 motivates the problem: standard metrics hide differences in instance difficulty.
- Section 2 introduces IRT as a latent-variable language for ability, difficulty, and discrimination.
- Section 3 presents Beta4-IRT and connects it to CLAIRE, where model agreement generates the response matrix.


## Practical Notes

- All notebooks use toy problems so they can be understood quickly in class.
- The activity notebooks are intentionally incomplete and include prompts for discussion.
- The CLAIRE notebook follows the workflow described in the paper, but keeps the classroom computation lightweight.
