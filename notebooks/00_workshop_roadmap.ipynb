# Bristol Roadmap

## Suggested Session Timing

Each section follows the same rhythm:

- **15 minutes** of theory and motivation
- **15 minutes** of guided hands-on explanation
- **30 minutes** for participants to solve the activity notebook

## Teaching Arc

The workshop now progresses from ordinary supervised evaluation to latent-variable reasoning, then to bounded-response Beta4 modeling, then back to unsupervised instance difficulty, and finally to the full CLAIRE workflow.

## Practical Notes

- Keep the answer notebooks as consolidation material after each activity.
- Use `05_01_extra_negative_disc.ipynb` as a companion analysis rather than the main participant task.
- The site mirrors the same five-section order.
